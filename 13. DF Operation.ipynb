```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
# Filter Data Frame

# Get all records where all records have tips less than $3

df_filter = df[df['tip'] < 3.0]
df_filter
```

```python
print("shape of original dataframe",df.shape)
print("shape of filtered dataframe",df_filter.shape)
```

```python
# Get all records where all records have tips less than $3 and sex is Male

df_filter_Gender = df[(df['tip'] < 3.0) & (df['sex']=='Male')]
print("shape of original dataframe",df.shape)
print("shape of filtered dataframe",df_filter_Gender.shape)
```

```python
# Select the appropriate columns
print("Columns in original dataframe :",df.columns)

#If we need only "total_bill", "tip" and "day" for our analysis

df_select = df[['total_bill', 'tip','day']]

print("Columns in dataframe after select:",df_select.columns)
```

```python
# Filter and select together

df_filter_select = df.loc[df['tip'] < 3.0,['total_bill', 'tip','day']]
df_filter_select
```

```python
#Summarization

import pandas as pd
import numpy as np

df = pd.DataFrame([[1.4, np.nan], [7.1, -4.5],
                [np.nan, np.nan], [0.75, -1.3]],
               index=['a', 'b', 'c', 'd'],
               columns=['one', 'two'])

print("Original Data Frame is: \n",df)

print("Sum is: \n",df.sum())

print("Sum over the axis 1 : \n",df.sum(axis = 1))
```

```python
#NA values are excluded unless the entire slice (row or column in this case) is NA. This
#can be disabled using the skipna option:
print("Original Data Frame is: \n",df)
print("Sum with NA: \n",df.sum(axis=1, skipna=False))
print("Mean with NA: \n",df.mean(axis=1, skipna=False))
```

# NA and None Values

Pandas automatically type-casts when NA values are present. For example, if we set a value in an integer array to np.nan, it will automatically be upcast to a floating-point type to accommodate the NA:

```python
x = pd.Series(range(2), dtype=int)
x
```

```python
x[0] = None
x
```

# Typeclass	    Conversion When Storing NAs	    NA Sentinel Value

#floating	          No change	                                         np.nan
#object	              No change	                                         None or np.nan
#integer	          Cast to float64	                                 np.nan
#boolean	          Cast to object	                                 None or np.nan


Keep in mind that in Pandas, string data is always stored with an object dtype.

```python
# Merge

import pandas as pd

df1 = pd.DataFrame({'key': ['b', 'b', 'a', 'c', 'a', 'a', 'b'],
                 'data1': range(7)})

df2 = pd.DataFrame({'key': ['a', 'b', 'd'],
                 'data2': range(3)})

print("First Dataframe: \n",df1)

print("Second Dataframe: \n",df2)

#This is an example of a many-to-one merge situation; the data in df1 has multiple rows labeled a and b, 
#whereas df2 has only one row for each value in the key column. Calling merge with these objects we obtain:

print("Many-to-one Merge: \n",pd.merge(df1, df2))
```

```python
print("Many-to-one Merge: \n",pd.merge(df1, df2, on = 'key'))
```

```python
print("Left Outer Join: \n",pd.merge(df1, df2, on='key', 
                                     how='left'))
```
